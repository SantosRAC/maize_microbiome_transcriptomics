```python
import pandas as pd
```

```python
# Read the file and create a dataframe
df_hisat2_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/wholegenomemapping_hisat2_percent.txt',
                                    header=None, names=['percent_alignment'])
df_sortmerna_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/rrna_kremling2018_alignment_percent.txt',
                                    header=None, names=['percent_alignment'])
df_kraken2maize_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/kraken2_kremling2018_maize_mapping_percent.txt',
                                       header=None, names=['percent_alignment'])
df_kraken2uncl_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/kraken2_kremling2018_uncl_mapping_percent.txt',
                                       header=None, names=['percent_alignment'])
df_salmon2zma2_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/Zma2_salmon_mapping_reprtranscripts.txt',
                                       header=None, names=['percent_alignment'])
df_salmon2zma2fullset_kremling = pd.read_csv('/home/rsantos/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/Zma2_salmon_mapping_fulltranscripts.txt',
                                       header=None, names=['percent_alignment'])

# Generate a boxplot
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Generate boxplots for each vector of values
ax.boxplot([df_hisat2_kremling['percent_alignment'],
            df_sortmerna_kremling['percent_alignment'],
            df_kraken2maize_kremling['percent_alignment'],
            df_kraken2uncl_kremling['percent_alignment'],
            df_salmon2zma2_kremling['percent_alignment'],
            df_salmon2zma2fullset_kremling['percent_alignment']])

# Set the x-axis tick labels
ax.set_xticklabels(['HISAT2 (Maize v5)',
                    'SortMERna',
                    'Kraken2 Maize',
                    'Kraken2 Unclassified',
                    'Salmon (Maize v5)',
                    'Salmon (Maize v5 - full set of transcripts)'], rotation=75)

# Set the y-axis label
ax.set_ylabel('Percent Alignment')

# Set the title
ax.set_title('Boxplots of Percent Alignment')

# Show the plot
plt.show()
```
