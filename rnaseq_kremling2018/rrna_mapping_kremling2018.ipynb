```python
import pandas as pd
```

```python
# Read the file and create a dataframe
df_sortmerna_kremling = pd.read_csv('/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/rrna_kremling2018_alignment_percent.txt',
                                    header=None, names=['percent_alignment'])
df_kraken2maize_kremling = pd.read_csv('/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/kraken2_kremling2018_maize_mapping_percent.txt',
                                       header=None, names=['percent_alignment'])
df_kraken2uncl_kremling = pd.read_csv('/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/kraken2_kremling2018_uncl_mapping_percent.txt',
                                       header=None, names=['percent_alignment'])
df_salmon2zma_kremling = pd.read_csv('/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/Zma_salmon_mapping.txt',
                                       header=None, names=['percent_alignment'])
df_salmon2zma2_kremling = pd.read_csv('/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/rnaseq_kremling2018/Zma2_salmon_mapping.txt',
                                       header=None, names=['percent_alignment'])

# Generate a boxplot
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Generate boxplots for each vector of values
ax.boxplot([df_sortmerna_kremling['percent_alignment'],
            df_kraken2maize_kremling['percent_alignment'],
            df_kraken2uncl_kremling['percent_alignment'],
            df_salmon2zma_kremling['percent_alignment'],
            df_salmon2zma2_kremling['percent_alignment']])

# Set the x-axis tick labels
ax.set_xticklabels(['SortMERna',
                    'Kraken2 Maize',
                    'Kraken2 Unclassified',
                    'Salmon (Maize v4)',
                    'Salmon (Maize v5)'], rotation=45)

# Set the y-axis label
ax.set_ylabel('Percent Alignment')

# Set the title
ax.set_title('Boxplots of Percent Alignment')

# Show the plot
plt.show()
```
