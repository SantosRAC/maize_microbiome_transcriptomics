# Plotting Module sizes

Plotting number of OTUs and genes per module in day and night bipartite (cross-correlation) networks.

Modules were generated using the Beckett method with the bipartite networks for day and night for which correlations were generated using the same number of plots (shared by day and night samples, 174).

```python
import pandas as pd

# Define two lists for day
day_gene_count = [6, 15, 29, 55, 113, 21, 57]
day_otu_count = [5, 28, 21, 3, 6, 7, 7]

# Create a dataframe from the two lists
day_df = pd.DataFrame({'Gene': day_gene_count, 'OTU': day_otu_count})
print(day_df)

# Define two lists for night
night_gene_count = [76, 92, 64, 38, 203, 176, 50]
night_otu_count = [11, 5, 20, 6, 5, 2, 34]

# Create a dataframe from the two lists
night_df = pd.DataFrame({'Gene': night_gene_count, 'OTU': night_otu_count})
print(night_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the width of the bars
bar_width = 0.35

# Define the positions of the bars
r1 = np.arange(1, len(day_df)+1)
r2 = [x + bar_width for x in r1]

# Create the bar plot
plt.figure(figsize=(10, 5))
plt.bar(r1, day_df['Gene'], color='blue', width=bar_width, edgecolor='grey', label='Gene')
plt.bar(r2, day_df['OTU'], color='red', width=bar_width, edgecolor='grey', label='OTU')

# Add xticks on the middle of the group bars
plt.xlabel('Modules', fontweight='bold')
plt.xticks([r for r in range(1, len(day_df)+1)])

# Add labels and title
plt.ylabel('Count')
plt.title('Counts in Day Modules')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
