# Cross-correlation analyses



```python
import pandas as pd
```

```python
sparxcc_day_signpairs = pd.read_csv("/home/santosrac/Projects/UGA_RACS/IntegrationMetataxExpression/SparXCC_OTUs_pag2025/SparXCC_output_day_pag25_edge_list.txt",
            sep="\t")
sparxcc_night_signpairs = pd.read_csv("/home/santosrac/Projects/UGA_RACS/IntegrationMetataxExpression/SparXCC_OTUs_pag2025/SparXCC_output_night_pag25_edge_list.txt",
            sep="\t")
```

```python
import matplotlib.pyplot as plt

# Separate positive and negative correlation coefficients
positive_day = sparxcc_day_signpairs[sparxcc_day_signpairs['CorrelationCoefficient'] > 0]
negative_day = sparxcc_day_signpairs[sparxcc_day_signpairs['CorrelationCoefficient'] < 0]

# Plot histogram for day data
plt.figure(figsize=(10, 5))
plt.hist(positive_day['CorrelationCoefficient'], bins=20, color='blue', alpha=0.7, label='Positive Day')
plt.hist(negative_day['CorrelationCoefficient'], bins=20, color='red', alpha=0.7, label='Negative Day')
plt.title('Histogram of Correlation Coefficients (Day)')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Separate positive and negative correlation coefficients
positive_night = sparxcc_night_signpairs[sparxcc_night_signpairs['CorrelationCoefficient'] > 0]
negative_night = sparxcc_night_signpairs[sparxcc_night_signpairs['CorrelationCoefficient'] < 0]

# Plot histogram for day data
plt.figure(figsize=(10, 5))
plt.hist(positive_night['CorrelationCoefficient'], bins=20, color='blue', alpha=0.7, label='Positive Night')
plt.hist(negative_night['CorrelationCoefficient'], bins=20, color='red', alpha=0.7, label='Negative Night')
plt.title('Histogram of Correlation Coefficients (Night)')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
