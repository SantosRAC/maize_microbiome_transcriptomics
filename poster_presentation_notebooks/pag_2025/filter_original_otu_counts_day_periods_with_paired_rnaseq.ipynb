# Analysis with the OTUs table

Separating OTUs for day and night samples, to analyze the cross-correlations.

This notebook also describes filtering steps, which is the same used for the Plant Center Retreat (OTUs are kept if 50% of samples have at least 0.001 relative abundance; also, OTUs with coefficient of variation above the 0.25 quartile is selected for correlations).

```python
import pandas as pd

otu_table_df = pd.read_table('/home/santosrac/Projects/UGA_RACS/16S/otu_matrices/original_counts/2f_otu_table.sample_filtered.no_mitochondria_chloroplast.tsv',
                                      comment='#', dtype = {'OTU': str})

otu_table_df.set_index('OTU_ID', inplace=True)
otu_table_df.head()
```

```python
# Day samples
otu_table_day_df = otu_table_df.filter(like='LMAD')
otu_table_day_df.head()
```

```python
# Night samples
otu_table_night_df = otu_table_df.filter(like='LMAN')
otu_table_night_df.head()
```

```python
# Renaming columns of the day OTU dataframe
column_mapping_d = {col: col.split(".")[2]+"_"+col.split(".")[1] for col in otu_table_day_df.columns}
otu_table_day_df.rename(columns=column_mapping_d, inplace=True)
otu_table_day_df.head()
```

```python
# Renaming columns of the night OTU dataframe
column_mapping_n = {col: col.split(".")[2]+"_"+col.split(".")[1] for col in otu_table_night_df.columns}
otu_table_night_df.rename(columns=column_mapping_n, inplace=True)
otu_table_night_df.head()
```

## Generating CPM file to import in CoNekT Grasses Microbiome

To use in CoNekT Grasses Microbiome, the original OTU counts will be normalized to CPM.

A first step is to filter out microbiome samples that have no associated RNA-seq.

Importing the run annotation file:

```python
otu_annotation_df = pd.read_table('/home/santosrac/Projects/UGA_RACS/CoNekT_Grasses_Microbiome/datasets/Maize/microbiome_runs_annotation.txt')
otu_annotation_df.set_index('RunID', inplace=True)
otu_annotation_df.head()
```

```python
otu_day_2filter = otu_table_day_df.copy()
column_mapping_d = {col: "microbiome_day_"+col for col in otu_table_day_df.columns}
otu_day_2filter.rename(columns=column_mapping_d, inplace=True)
otu_day_2filter.head()
```

```python
otu_night_2filter = otu_table_night_df.copy()
column_mapping_n = {col: "microbiome_night_"+col for col in otu_table_night_df.columns}
otu_night_2filter.rename(columns=column_mapping_n, inplace=True)
otu_night_2filter.head()
```

```python
concatenated_day_night_df = pd.concat([otu_day_2filter, otu_night_2filter], axis=1)
print(concatenated_day_night_df.shape)
```

```python
concatenated_day_night_filtered_df = concatenated_day_night_df[concatenated_day_night_df.columns.intersection(otu_annotation_df.index)]
concatenated_day_night_filtered_df.head()
```

```python
from bioinfokit.analys import norm

# CPM (or RPM, reads per million) normalization
# Following Renesh Bedre's blog:
# https://www.reneshbedre.com/blog/expression_units.html#rpm-or-cpm-reads-per-million-mapped-reads-or-counts-per-million-mapped-reads-
nm = norm()
nm.cpm(df=concatenated_day_night_filtered_df)
otu_concatenated_day_night_filtered_cpm = nm.cpm_norm
```

```python
otu_concatenated_day_night_filtered_cpm.head()
```

```python
otu_concatenated_day_night_filtered_cpm.to_csv('/home/santosrac/Repositories/maize_microbiome_transcriptomics/poster_presentation_notebooks/pag_2025/otu_concatenated_day_night_filtered_cpm.tsv', sep='\t')
```

## Filtering the OTU tables based on the RNA-Seq (paired) samples

```python
expression_counts_day = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/correlations_rnaseq_metataxonomics/kremling_expression_v5_day.tsv",
                                    sep="\t")
expression_counts_day.set_index("Name", inplace=True)
expression_counts_night = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/correlations_rnaseq_metataxonomics/kremling_expression_v5_night.tsv",
                                    sep="\t")
expression_counts_night.set_index("Name", inplace=True)
```

```python
# Filter columns of otu_table_day_df based on columns of expression_counts_day
filtered_otu_table_day_df = otu_table_day_df[otu_table_day_df.columns.intersection(expression_counts_day.columns)]
filtered_otu_table_day_df.head()
```

```python
# Filter columns of otu_table_night_df based on columns of expression_counts_night
filtered_otu_table_night_df = otu_table_night_df[otu_table_night_df.columns.intersection(expression_counts_night.columns)]
filtered_otu_table_night_df.head()
```

```python
print(filtered_otu_table_day_df.shape)
print(filtered_otu_table_night_df.shape)
```

```python
filtered_otu_table_day_df.columns
```

```python
# Renaming columns before merging again
column_mapping_d_filtered = {col: col+"_d" for col in filtered_otu_table_day_df.columns}
filtered_otu_table_day_df.rename(columns=column_mapping_d_filtered, inplace=True)
filtered_otu_table_day_df.head()
```

```python
# Renaming columns before merging again
column_mapping_n_filtered = {col: col+"_n" for col in filtered_otu_table_night_df.columns}
filtered_otu_table_night_df.rename(columns=column_mapping_n_filtered, inplace=True)
filtered_otu_table_night_df.head()
```

```python
# Concatenate the two dataframes before filtering
filtered_otu_table_day_night_df = pd.concat([filtered_otu_table_day_df, filtered_otu_table_night_df], axis=1)
filtered_otu_table_day_night_df.shape
```

Computing relative abundance for the dataframe (day and night samples that have a pair in the RNA-Seq data)

```python
filtered_otu_table_day_night_rel_abund = filtered_otu_table_day_night_df.divide(filtered_otu_table_day_night_df.sum())
filtered_otu_table_day_night_rel_abund = filtered_otu_table_day_night_rel_abund * 100
filtered_otu_table_day_night_rel_abund.head()
```

Filtering OTUs based on relative abundance across samples:

```python
otus_tokeep = filtered_otu_table_day_night_rel_abund[(filtered_otu_table_day_night_rel_abund > 0.001).sum(axis=1) >= (filtered_otu_table_day_night_rel_abund.shape[1] * 0.5)].index
```

```python
filtered_otu_table_day_night_filtered_rel_abund = filtered_otu_table_day_night_df[filtered_otu_table_day_night_df.index.isin(otus_tokeep)]
filtered_otu_table_day_night_filtered_rel_abund.shape
```

Filtering OTUs based on the quartile value of coefficient of variation:

```python
import numpy as np
filtered_otu_table_day_night_filtered_rel_abund_cv = filtered_otu_table_day_night_filtered_rel_abund.apply(lambda row: np.std(row) / np.mean(row), axis=1)
filtered_otu_table_day_night_filtered_rel_abund_cv_filtered = filtered_otu_table_day_night_filtered_rel_abund.loc[filtered_otu_table_day_night_filtered_rel_abund_cv[filtered_otu_table_day_night_filtered_rel_abund_cv > filtered_otu_table_day_night_filtered_rel_abund_cv.quantile(q=0.25)].index]
```

```python
filtered_otu_table_day_night_filtered_rel_abund_cv_filtered.shape
```

```python
filtered_otu_table_day_filtered_rel_abund_cv_filtered = filtered_otu_table_day_night_filtered_rel_abund_cv_filtered.filter(like='_d')
print(filtered_otu_table_day_filtered_rel_abund_cv_filtered.shape)
filtered_otu_table_night_filtered_rel_abund_cv_filtered = filtered_otu_table_day_night_filtered_rel_abund_cv_filtered.filter(like='_n')
print(filtered_otu_table_night_filtered_rel_abund_cv_filtered.shape)
```

```python
column_mapping_d_final = {col: col.replace("_d", "") for col in filtered_otu_table_day_filtered_rel_abund_cv_filtered.columns}
filtered_otu_table_day_filtered_rel_abund_cv_filtered.rename(columns=column_mapping_d_final, inplace=True)
column_mapping_n_final = {col: col.replace("_n", "") for col in filtered_otu_table_night_filtered_rel_abund_cv_filtered.columns}
filtered_otu_table_night_filtered_rel_abund_cv_filtered.rename(columns=column_mapping_n_final, inplace=True)
```

```python
print(filtered_otu_table_night_filtered_rel_abund_cv_filtered.shape)
print(filtered_otu_table_day_filtered_rel_abund_cv_filtered.shape)
```

```python
filtered_otu_table_night_filtered_rel_abund_cv_filtered.to_csv("filtered_otu_table_night_filtered_rel_abund_cv_filtered.tsv", sep="\t")
filtered_otu_table_day_filtered_rel_abund_cv_filtered.to_csv("filtered_otu_table_day_filtered_rel_abund_cv_filtered.tsv", sep="\t")
```
