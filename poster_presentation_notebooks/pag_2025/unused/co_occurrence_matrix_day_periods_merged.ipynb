# Generating a matrix with merged day and night OTU counts for the co-occurrence network

We analyzed the cross-correlation between host genes and microbiome. For that purpose, we previously filtered out OTUs with low variance and low abundance. In this notebook, I (RACS) will generate a matrix with merged reads for each plot (day and night merged), to construct a co-occurrence network for the manuscript main text. The idea is also to compare and make sure this network is very similar to the others were analyzed.

```python
import pandas as pd

merged_otu_df = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/16S_wallace2018/combine_day_night_samples/summed_day_night_otu_counts.tsv",
            sep='\t', index_col=0, dtype={'OTU_ID': str})
merged_otu_df.head()
```

```python
import pandas as pd

summed_otus_filtered_crosscor = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/poster_presentation_notebooks/pag_2025/summed_otu_counts_separate_day_night_filtered_cv_filtered_no_duplicated.tsv",
            sep='\t', index_col=0, dtype={'OTU_ID': str})
summed_otus_filtered_crosscor.head()
```

```python
import pandas as pd

original_otus_filtered = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/poster_presentation_notebooks/pag_2025/otu_table_sample_filtered_day_otu_filtered.tsv",
            sep='\t', index_col=0, dtype={'OTU_ID': str})
original_otus_filtered.head()
```
