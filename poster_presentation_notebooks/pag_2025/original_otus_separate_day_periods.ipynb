# Analysis with the OTUs table

Separating OTUs for day and night samples, to analyze the cross-correlations.

```python
import pandas as pd

otu_table_df = pd.read_table('/home/santosrac/Projects/UGA_RACS/16S/otu_matrices/original_counts/2f_otu_table.sample_filtered.no_mitochondria_chloroplast.tsv',
                                      comment='#', dtype = {'OTU': str})

otu_table_df.set_index('OTU_ID', inplace=True)
otu_table_df.head()
```

Computing relative abundance to filter OTUs

```python
# Compute relative abundance
otu_rel_abund = otu_table_df.divide(otu_table_df.sum())
otu_rel_abund = otu_rel_abund * 100
```

```python
# Filter out OTUs with less than 0.1% relative abundance in at least 50% of the samples
otus_tokeep = otu_rel_abund[(otu_rel_abund > 0.001).sum(axis=1) >= (otu_rel_abund.shape[1] * 0.5)].index
len(otus_tokeep)
```

```python
# Day samples
otu_table_day_df = otu_table_df.filter(like='LMAD')
otu_table_day_df.head()
```

```python
# Night samples
otu_table_night_df = otu_table_df.filter(like='LMAN')
otu_table_night_df.head()
```

```python
# Renaming columns of the day OTU dataframe
column_mapping_d = {col: col.split(".")[2]+"_"+col.split(".")[1] for col in otu_table_day_df.columns}
otu_table_day_df.rename(columns=column_mapping_d, inplace=True)
otu_table_day_df.head()
```

```python
# Renaming columns of the night OTU dataframe
column_mapping_n = {col: col.split(".")[2]+"_"+col.split(".")[1] for col in otu_table_night_df.columns}
otu_table_night_df.rename(columns=column_mapping_n, inplace=True)
otu_table_night_df.head()
```

Filtering the OTU tables based on the RNA-Seq (paired) samples

```python
expression_counts_day = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/correlations_rnaseq_metataxonomics/kremling_expression_v5_day.tsv",
                                    sep="\t")
expression_counts_day.set_index("Name", inplace=True)
expression_counts_night = pd.read_csv("/home/santosrac/Repositories/maize_microbiome_transcriptomics/correlations_rnaseq_metataxonomics/kremling_expression_v5_night.tsv",
                                    sep="\t")
expression_counts_night.set_index("Name", inplace=True)
```

```python
# Filter columns of otu_table_day_df based on columns of expression_counts_day
filtered_otu_table_day_df = otu_table_day_df[otu_table_day_df.columns.intersection(expression_counts_day.columns)]
filtered_otu_table_day_df.head()
```

```python
# Filter columns of otu_table_night_df based on columns of expression_counts_night
filtered_otu_table_night_df = otu_table_night_df[otu_table_night_df.columns.intersection(expression_counts_night.columns)]
filtered_otu_table_night_df.head()
```

```python
# Exporting filtered OTU tables
filtered_otu_table_day_df.to_csv("otu_table_sample_filtered_day.tsv",
                                sep="\t")
filtered_otu_table_night_df.to_csv("otu_table_sample_filtered_night.tsv",
                                sep="\t")
```
