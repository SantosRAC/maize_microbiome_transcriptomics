## ASV data (taxonomy and feature table)

Importing the ASV table and the taxonomic assignments (q20, 150bp, forward):

```python
import pandas as pd

asv_taxonomy_df = pd.read_table('/media/rsantos/4TB_drive/Projects/UGA_RACS/16S/Qiime2/taxonomic_bar_plots/q20_150bp/silva_taxonomy_fw_q20_150bp/taxonomy.tsv')
asv_taxonomy_df.set_index('ASV', inplace=True)
print(asv_taxonomy_df.head())

asv_featuretable_df = pd.read_table('/media/rsantos/4TB_drive/Projects/UGA_RACS/16S/Qiime2/dada2/q20_150bp/table-paired-end_wallace2018_assingle_forward_q20_150bp-dada2/featuretable_fw_q20_150bp.tsv')
asv_featuretable_df.set_index('ASV', inplace=True)
print(asv_featuretable_df.head())
```

```python
asv_featuretable_unassigned_df = asv_featuretable_df.loc[asv_taxonomy_df[asv_taxonomy_df['Taxon'] == 'Unassigned'].index]
print('Total ASV frequencies', asv_featuretable_df.sum().sum())
print('ASV frequencies for ASVs with unassigned taxon', asv_featuretable_unassigned_df.sum().sum())
```

## OTU data (table with taxonomy attached)

Importing the OTU table with the taxonomic assignments (original .biom, after exporting the table with biom)

```python
# Created with qiime2 2023.9 using the following command:
# biom convert -i 2f_otu_table.sample_filtered.no_mitochondria_chloroplast.biom -o 2f_otu_table.sample_filtered.no_mitochondria_chloroplast_taxonomy.tsv --to-tsv --header-key taxonomy
```

```python
otu_table_df = pd.read_table('/media/rsantos/4TB_drive/Projects/UGA_RACS/16S/Workflow/2_QiimeOtus/2f_otu_table.sample_filtered.no_mitochondria_chloroplast_taxonomy.tsv')
otu_table_df.set_index('OTU', inplace=True)
print(otu_table_df.head())
```

```python
sra_run_table_16s = '/home/rsantos//Repositories/maize_microbiome_transcriptomics/16S_wallace2018/SraRunInfo_Wallace_etal_2018.csv'
sra2sampleid = {}

with open(sra_run_table_16s, 'r') as file:

    _ = file.readline()

    for line in file:
        fields = line.strip().split(',')
        metataxonomics_run_id = fields[0]
        sra2sampleid[fields[11]]=metataxonomics_run_id
```

```python
otu_table_df = otu_table_df.rename(columns=sra2sampleid)
otu_table_df.columns = [str(x) for x in otu_table_df.columns]
otu_table_df.index = [str(x) for x in otu_table_df.index]
```
