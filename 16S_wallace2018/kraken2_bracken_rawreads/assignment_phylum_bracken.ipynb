Here, I (RACS) plots the Bracken results after running Kraken2 with the 16S raw reads, using the special database for 16S (SILVA). Bracken results are plotted at the phylum level, after running the Bracken script `combine_bracken_output.py` with `.bracken` files, without any filtering of confidence score when running Kraken2.

```python
import pandas as pd

bracken_rawreads_no_confscore_P_df = pd.read_table('/media/renato/SSD1TB/Projects/UGA_RACS/16S/Bracken/no_confidence_Score/P_kraken2_out_bracken/merge_fraq.txt')
bracken_rawreads_no_confscore_P_df.drop(['taxonomy_id', 'taxonomy_lvl'], axis=1, inplace=True)
bracken_rawreads_no_confscore_P_df.set_index('name', inplace=True)

# Display the dataframe
bracken_rawreads_no_confscore_P_df.head()
```

```python
import matplotlib.pyplot as plt
bracken_rawreads_P_df_transposed = bracken_rawreads_no_confscore_P_df.loc[['Proteobacteria', 'Bacteroidota', 'Actinobacteriota', 'Firmicutes']].T
bracken_rawreads_P_df_transposed.plot(kind='bar', stacked=True, width=0.6)
plt.xticks([])
plt.show()
```

```python
sra_run_table_16s = '/media/renato/SSD1TB/Repositories/maize_microbiome_transcriptomics/16S_wallace2018/SraRunInfo_Wallace_etal_2018.csv'
#sra_run_table_16s = '/home/rsantos/Repositories/maize_microbiome_transcriptomics/16S_wallace2018/SraRunInfo_Wallace_etal_2018.csv'
sra2sampleid = {}

with open(sra_run_table_16s, 'r') as file:

    _ = file.readline()

    for line in file:
        fields = line.strip().split(',')
        metataxonomics_run_id = fields[0]
        sra2sampleid[fields[11]]=metataxonomics_run_id
```

```python
otu_table_df = otu_table_df.rename(columns=sra2sampleid)
otu_table_df.columns = [str(x) for x in otu_table_df.columns]
otu_table_df.index = [str(x) for x in otu_table_df.index]
```
