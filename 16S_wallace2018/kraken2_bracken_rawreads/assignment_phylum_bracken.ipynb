Here, I (RACS) plots the Bracken results after running Kraken2 with the 16S raw reads, using the special database for 16S (SILVA). Bracken results are plotted at the phylum level, after running the Bracken script `combine_bracken_output.py` with `.bracken` files, without any filtering of confidence score when running Kraken2.

```python
import pandas as pd

bracken_rawreads_P_df = pd.read_table('/media/renato/SSD1TB/Projects/UGA_RACS/16S/Bracken/P_kraken2_out_bracken/merge_fraq.txt')
bracken_rawreads_P_df.drop(['taxonomy_id', 'taxonomy_lvl'], axis=1, inplace=True)
bracken_rawreads_P_df.set_index('name', inplace=True)

# Display the dataframe
bracken_rawreads_P_df.head()
```
